# **1.What are Type I and II errors in hypothesis testing, and how do they impact decision-making?**
Ans:  
  **Type I and Type II Errors in Hypothesis Testing**  
  In hypothesis testing, we test two statements:  
- **H₀ (Null Hypothesis)** – No effect, no difference.  
- **H₁ (Alternative Hypothesis)** – There is an effect or difference.   
While making decisions, we may make mistakes. These mistakes are called Type I and Type II errors.
  **Type I Error (False Positive)**   
  **Definition:**   
  Rejecting the null hypothesis when it is actually true.   
  **Meaning:**      
  You think there is an effect, but actually there is none.   
  **Example:**    
  A COVID test says positive, but the person is actually healthy.   
  **Symbol:**   
 (alpha) – usually set at 0.05 → 5% chance of Type I error.   
  **Impact on Decision-Making:**    
  1.Increases false alarms.
  2.We may take unnecessary actions.  
  3.Risky when the cost of a false positive is high (e.g., diagnosing a disease incorrectly).   
**Type II Error (False Negative)**  
  **Definition:**   
  Failing to reject the null hypothesis when it is actually false.    
  **Meaning:**    
  You think there is no effect, but actually there is one.  
  Example:    
  A COVID test says negative, but the person is actually infected.    
  **Symbol:**   
  β (beta).   
  **Impact on Decision-Making:**    
  - Important signals are missed.   
  - May result in no action when action was needed.
  - Dangerous when missing a real effect has high cost (e.g., failing to detect a defect in a factory machine).     
  **Summary Table**   
  Error Type-----------What Happens-----------Real-Life-----------Example Risk  
  Type I Error-------|False Positive|--|Healthy person markedas sick|--|Incorrectactivation / false alarm|                           
  Type II Error------|False Negative|--|Sick person marked as healthy|--|Missed detection / delayed action|   
  **How They Impact Decision-Making**     
  - Lower α (Type I error) → fewer false alarms, but increases chances of - - missing real effects (more Type II errors).
  - Decrease Type II error → increases sensitivity, but may raise Type I errors.
  - Choosing α and β depends on the situation, cost, and risk.



# **2.What is the p-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?**
Ans.
  **1. Meaning of P-value (Introduction)**    
  In hypothesis testing, the p-value is one of the most important statistical measures used to make decisions about the null hypothesis. It helps us understand whether the observed data is consistent with the assumption that the null hypothesis (H₀) is true.    
  **A p-value is defined as:**  
  **The probability of obtaining the observed results, or more extreme results, assuming that the null hypothesis is true.**    
  In simple terms, it measures how surprising the data is under H₀.   
  **2. Role of Null and Alternative Hypothesis**
  - **Null Hypothesis (H₀):** Indicates no significant effect, no difference, or no relationship.     
  Example: “The new teaching method has no effect on student marks.       
  - **Alternative Hypothesis (H₁):** Indicates a significant effect or difference.      
  Example: “The new teaching method improves student marks.”    
  The p-value helps us choose between these two hypotheses based on evidence from the sample data.    
  **3. How P-value Is Calculated (Conceptual Explanation)**     
  Although the exact calculation depends on the test (t-test, z-test, chi-square, etc.), the p-value essentially compares:    
  - the observed test statistic, and      
  - the distribution expected under H₀.   
  Higher differences between observed and expected values produce smaller p-values, meaning stronger evidence against H₀.     
  **4. Interpretation of P-value**    
  **(a) When p-value ≤ α (Significance Level) **  
  Most commonly α = 0.05 or 0.01.   
  If p-value is less than or equal to α:    
  - Evidence against H₀ is strong.    
  - The observed result is unlikely under H₀.   
  - We reject the null hypothesis.    
  - The result is statistically significant.    
  Example:    
  p = 0.03, α = 0.05 → 0.03 ≤ 0.05 → Reject H₀    
  Interpretation: There is significant evidence that the effect is real.
  **(b) When p-value > α**    
  If the p-value is greater than α:   
  - Evidence against H₀ is weak.    
  - The observed data is consistent with the null hypothesis.
  - We fail to reject the null hypothesis.    
  - The result is not statistically significant.  
  Example:  
  p = 0.42, α = 0.05 → 0.42 > 0.05 → Fail to Reject H₀  
  Interpretation: There is no significant evidence of an effect.    
  **5. Practical Examples**     
  **(A) Medical Example**   
  H₀: A new drug has no effect.     
  Collected data gives p = 0.02.    
  Since 0.02 < 0.05 → Reject H₀ →   
  Conclusion: The drug is effective.    
  **(B) Quality Control Example**   
  H₀: Machine is working normally.    
  P-value = 0.35.   
  Since 0.35 > 0.05 → Fail to Reject H₀ →   
  Conclusion: Machine performance is normal.      
  **(C) Education Example**   
  H₀: New teaching method does not improve scores.    
  p = 0.07.   
  Since 0.07 > 0.05 → Fail to Reject H₀ →     
  Conclusion: No significant improvement.   
  **6. Relationship with Type I and Type II Errors**    
  - A small p-value reduces the chance of Type II error (false negative).     
  - Choosing α = 0.05 defines how much risk we accept for Type I error (false positive).    
  Thus, the p-value plays an important role in balancing statistical errors.  
  **7. Misinterpretations to Avoid (Important in Exams)**   
  - A high p-value does not prove H₀ true; it only means lack of evidence to reject it.   
  - A low p-value does not measure effect size; it only shows evidence strength.
  - The p-value is not the probability that H₀ is true.


```python
#Python Code for Hypothesis Test (T-test)
import numpy as np
from scipy.stats import ttest_1samp

# Sample data
weights = np.array([198, 202, 199, 201, 200, 197, 203, 200, 199, 201, 204, 198])

# Null hypothesis mean (claimed mean)
population_mean = 200

# Perform one-sample t-test
t_stat, p_value = ttest_1samp(weights, population_mean)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

# **3. Explain the difference between a Z-test and a T-test, including when to use each.**
  Ans.**1. Introduction**           
  In statistics, both the Z-test and T-test are hypothesis tests used to compare sample data against population parameters or to compare two groups. They help determine whether differences observed in data are statistically significant.          
  Although both tests follow similar logic, they are used under different conditions.           
  **2. Z-test**             
  **Definition:**                 
  A Z-test is a statistical test used when the population variance (σ²) or population standard deviation (σ) is known, and the sample size is large.
  When to Use Z-test:           
  You should use a Z-test when:
  1.  	Sample size is large (n ≥ 30).
  2.  	Population standard deviation (σ) is known.
  3.  	Data follows a normal distribution or the sample is large enough for the Central Limit Theorem to apply.
  4.  	Used for comparing:
  - A sample mean with population mean
  o    Two population proportions
  o    Two sample proportions
  **Examples:**
  - Testing if the average height of students differs from a known population mean.
  - Testing the difference between two population proportions in surveys.   
  **3. T-test**       
  **Definition:**           
  A T-test is used when the population standard deviation is unknown and the sample size is small. Instead of σ, we use the sample standard deviation (s).
  When to Use T-test:
  You should use a T-test when:
  1.  	Sample size is small (n < 30).
  2.  	Population standard deviation (σ) is unknown.
  3.  	Data is approximately normally distributed.
  4.  	Used for comparing:
  - One sample mean with population mean (One-sample t-test)
  o    Two sample means (Independent t-test)
  o    Paired observations (Paired t-test)
  Examples:
  - Testing if the average marks of 15 students differ from 50.
  - Comparing average weight before and after a diet program (paired t-test).   
  **4. Key Differences Between Z-test and T-test**      
  **Basis**---------------------------------------**Z-test**-------------------------------------**T-test**   
  Sample Size----------------------Large samples (n ≥ 30)----------------------Small samples (n < 30)  
  Population SD (σ)--------------------------Known-----------------------------------Unknown    
  Distribution Used--------------Standard normal(Z-distribution)------------Student’s t-distribution
  Shape of Curve------------------------Constant shape----------Shape changes with sample size (wider tails)    
  Degree of Freedom---------------------Not required----------------------------Required (n − 1)    
  When Used-----------------Estimate population mean or proportion-----Compare means when σ is unknown      
  Accuracy---------------------More accurate for large samples----------More accurate for small samples       
  **5. Why T-test Is Needed When n < 30?**               
  When the sample size is small, the sample standard deviation s is not a very accurate estimate of σ.
 The t-distribution adjusts for this uncertainty by having heavier tails, which increases reliability for small samples.            
  **6. Practical Example for Understanding**        
  Example 1: Z-test Scenario          
  A company knows its machine produces items with a known population SD = 3g.   
  A sample of 50 items is selected to test whether the mean weight differs from 200g.     
  - σ known → Z-test          
  - n ≥ 30 → Z-test         
  Example 2: T-test Scenario      
  A teacher wants to test whether the average marks of 20 students differ from 65.            
  Population SD is unknown.             
  - σ unknown → T-test              
  - n = 20 < 30 → T-test          
  **7. Conclusion**    
  The Z-test and T-test both serve the purpose of hypothesis testing, but the choice between them depends mainly on sample size and knowledge of population standard deviation.
  - Use Z-test when σ is known and sample is large.
  - Use T-test when σ is unknown or sample is small.    
  Choosing the correct test ensures the reliability and accuracy of statistical conclusions.

```python
#Python code for T-test
import numpy as np
from scipy.stats import ttest_1samp

# Sample data for bulb lifetime
sample = np.array([980, 1020, 1005, 995, 990, 985, 1010, 1008, 1002, 998,
                   1001, 1004, 975, 990, 995])

# Perform one-sample t-test
t_stat, p_value = ttest_1samp(sample, 1000)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
#Python Code for Z-test
import numpy as np
from scipy.stats import norm

# Given data
population_mean = 1000
population_std = 50
sample_mean = 990
n = 60

# Calculate Z-value
z_value = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# p-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z_value)))

print("Z-value:", z_value)
print("P-value:", p_value)
```

# **4.What is a confidence, and how does the margin of error influence its width and interpretation?**
Ans- **Confidence Interval (Theoretical Definition)**         
  A confidence interval (CI) is a range of values, calculated from sample data, that is likely to contain the true population parameter (such as mean or proportion).       
  It provides a measure of uncertainty and reliability of the estimate.
  - It does not guarantee the population parameter lies within the interval.
  - Instead, it indicates that if we repeat a study many times, a certain percentage (e.g., 95%) of the calculated intervals will contain the true population value.            
  **Margin of Error (Theoretical Definition)**            
  The margin of error is the maximum expected difference between the sample estimate and the true population parameter. It is added and subtracted from the sample estimate to form the confidence interval.        
  Formula (General Form):   
  Confidence Interval=Sample Estimate±Margin of Error         
  **How Margin of Error Influences the Width and Interpretation of CI**   
  The margin of error directly determines the width of the confidence interval:
  - Larger margin of error → Wider confidence interval    
            - Indicates more uncertainty in the estimate.
  - Smaller margin of error → Narrower confidence interval          
            - Indicates higher precision and more reliable estimate.
  **Factors Affecting Margin of Error**         
  Margin of error depends on:         
  Factor----------------------------------------Effect on CI            
  Sample size----------------------------Larger sample → smaller margin → narrower CI        
  Data variability-------------------------More variability → larger margin → wider CI        
  Confidence level (e.g., 95% → 99%)-----Higher confidence → larger margin → wider CI       
  **Interpretation**
  A confidence interval communicates two key points:      
  - It gives a range of plausible values for a population parameter.  
  - The chosen confidence level determines how sure we are that this interval contains the true parameter.    
  Example of interpretation (at 95% CI):
  “We are 95% confident that the true population mean lies within this interval.”

```python
import numpy as np
from scipy.stats import norm

# Sample data (student weights)
weights = np.array([55, 60, 62, 58, 54, 65, 63, 59, 57, 61,
                    56, 67, 64, 62, 59, 60, 55, 66, 63, 61,
                    58, 57, 64, 65, 62, 63, 60, 59, 61, 58,
                    57, 56, 60, 62, 65, 64, 63, 61, 59, 57,
                    56, 60, 62, 63, 64, 65, 59, 57, 58, 60])

# Sample statistics
mean = np.mean(weights)
std = np.std(weights, ddof=1)
n = len(weights)

# 95% confidence level Z-score
z_value = norm.ppf(0.975)

# Margin of Error
margin_of_error = z_value * (std / np.sqrt(n))

# Confidence Interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print("Sample Mean:", mean)
print("Margin of Error:", margin_of_error)
print(f"95% Confidence Interval: ({lower_bound}, {upper_bound})")
```

# **5.describe the purpose and assumptions of an ANOVA test. how does it extend hypothesis testing to more than two groups?**
Ans: **Purpose of ANOVA**   
  **ANOVA (Analysis of Variance) is a statistical test used to determine whether there are significant differences between the means of three or more independent groups**.     
  **Instead of comparing groups two at a time, ANOVA tests all group means simultaneously and tells us whether at least one group mean is different.**
  It is mainly used when:   
  - You want to compare more than two groups.     
  - You want to check whether a factor (e.g., teaching method, medicine type, fertilizer type) affects the outcome.       
  ANOVA works by comparing:     
  - **Between-group variance** → how different the group means are
 vs.
  - **Within-group variance** → how spread out the observations are inside each group
  If between-group variance is much larger, ANOVA concludes that group means are significantly different.     
  #  **Assumptions of ANOVA**          
  **ANOVA is valid only when the following assumptions are met:**
  **1. Independence of Observations**         
  Each observation must be independent. No individual should appear in more than one group.             
  Example: Students in different classes should not overlap.        
  **2. Normality**          
  Data in each group should be approximately normally distributed.    
  ANOVA is robust, so slight deviations are acceptable with large sample sizes.
  **3. Homogeneity of Variance (Equal Variances)**      
  All groups should have similar variances.       
  This is tested using Levene’s test or Bartlett’s test.        
  If variances are not equal, we use Welch’s ANOVA.         
  How ANOVA Extends Hypothesis Testing Beyond Two Groups    
  Traditional hypothesis tests like the t-test compare only two means.        
  **If we compare three or more groups using multiple t-tests:**    
  Probability of Type I Error increases     
  Results become unreliable     
  ANOVA solves this problem by testing all groups at the same time under one **hypothesis framework:**        
  **Null hypothesis (H₀):**       
  All group means are equal     
  μ₁ = μ₂ = μ₃ = … = μₖ   
  **Alternative hypothesis (H₁):**          
  At least one group mean is different        
  ANOVA does not tell which specific groups differ; it only tells if a difference exists.         
  If ANOVA is significant, researchers perform post-hoc tests (Tukey, Bonferroni, Scheffé) to identify the exact groups with differences.
  Example (Simple Explanation)          
  **A teacher wants to compare the marks of students taught with:**     
  Traditional method      
  Smart class method    
  Online method   
  ANOVA can test in a single step whether teaching method affects marks.    

```python
import numpy as np
from scipy.stats import f_oneway

# Exam scores for three teaching methods
method_A = np.array([78, 82, 75, 79, 81])
method_B = np.array([85, 80, 78, 82, 79])
method_C = np.array([88, 90, 86, 84, 89])

# One-way ANOVA Test
f_stat, p_value = f_oneway(method_A, method_B, method_C)

print("F-Statistic:", f_stat)
print("P-Value:", p_value)
```

# **6.Write a Python program to perform a one-simple Z-test and interpret the result for a given dataset.**
Ans:**Python Program: One-Sample Z-Test**         
Suppose a company claims that the average battery life of their phones is 10 hours.     
 You collected a sample of 30 phones to test this claim.
We will perform a one-sample Z-test to check whether the sample mean is significantly different from 10 hours.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
# Sample dataset (battery life in hours)
data = [9.5, 10.2, 9.8, 10.5, 9.9, 10.1, 9.7, 9.6, 10.4, 10.0,
        9.8, 10.3, 9.9, 10.1, 9.7, 10.2, 10.3, 9.8, 10.1, 9.6,
        10.4, 10.2, 9.9, 9.7, 10.0, 10.1, 9.8, 10.2, 9.9, 10.3]

# Step 1: Known population mean (claim)
mu_population = 10

# Step 2: Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=0)     # population SD (known or assumed)
n = len(data)

# Step 3: Calculate Z-statistic
z_value = (sample_mean - mu_population) / (sample_std / np.sqrt(n))

# Step 4: Calculate p-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z_value)))

# Print results
print("Sample Mean:", round(sample_mean, 3))
print("Sample Standard Deviation:", round(sample_std, 3))
print("Z-value:", round(z_value, 3))
print("P-value:", round(p_value, 4))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the Null Hypothesis — Sample mean is significantly different.")
else:
    print("Result: Fail to Reject the Null Hypothesis — No significant difference.")
```

# **Interpretation**        
Hypotheses
H₀ (Null): The true mean = 10 hours
H₁ (Alternative): The true mean ≠ 10 hours

# **8. Generate multiple samples from a non-normal distribution and implement the central limit Theorem using python.**

```python

import numpy as np
import matplotlib.pyplot as plt

# 1. Generate samples from a non-normal distribution (Exponential)
data = np.random.exponential(scale=1.0, size=5000)

plt.hist(data, bins=40, edgecolor="black")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Original Non-Normal Distribution (Exponential)")
plt.show()

# 2. Implement Central Limit Theorem

sample_means = []

# Generate 3000 samples, each of size 50
for _ in range(3000):
	sample = np.random.exponential(scale=1.0, size=50)
	sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)

plt.hist(sample_means, bins=40, edgecolor="black")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.title("Distribution of Sample Means (CLT Demonstration)")
plt.show()
```

# **9. Write a Python function to calculate and visualize the confidence interval for a sample mean.**
Ans- A confidence interval (CI) gives a range of values within which the true population mean is likely to lie.       
CI=xˉ±Zα/2×σnCI = \bar{x} \pm Z_{\alpha/2} \times \frac{\sigma}{\sqrt{n}}CI=xˉ±Zα/2​×n​σ​         
If population standard deviation is unknown, use sample standard deviation and a t-distribution.        
Purpose of CI       
  - To estimate the population mean with uncertainty.
  - Wider CI → More uncertainty
  - Narrow CI → Higher precision

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def confidence_interval_plot(sample, confidence=0.95):
	"""
	Calculate and visualize a confidence interval for the sample mean.
	:param sample: list or NumPy array of numeric data
	:param confidence: confidence level (default = 95%)
	"""

	# Convert to numpy array
	data = np.array(sample)

	# Sample statistics
	mean = np.mean(data)
	std = np.std(data, ddof=1)  	# sample standard deviation
	n = len(data)

	# Z-score for desired confidence level
	alpha = 1 - confidence
	z_value = norm.ppf(1 - alpha/2)

	# Margin of Error
	margin = z_value * (std / np.sqrt(n))

	# Confidence Interval
	lower = mean - margin
	upper = mean + margin

	print(f"Sample Mean: {mean}")
	print(f"{confidence*100}% Confidence Interval: ({lower}, {upper})")

	# Visualization
	plt.figure(figsize=(8, 4))

	plt.axvline(mean, linestyle='--', label="Sample Mean")
	plt.axvline(lower, color='red', label="Lower CI")
	plt.axvline(upper, color='green', label="Upper CI")

	plt.title(f"{int(confidence*100)}% Confidence Interval for Sample Mean")
	plt.xlabel("Value")
	plt.legend()
	plt.show()

	return (lower, upper)
```

**Explanation of the Code**       
**1. Convert data to NumPy**          
Ensures mathematical operations work efficiently.           
**2. Calculate Sample Mean and Standard Deviation**       
Used for estimating the population mean.      
**3. Compute Z-value**        
Used based on confidence level (e.g., 1.96 for 95%).

  This visually shows whether the mean is centered and how wide the interval is.

```python
import numpy as np
from scipy.stats import chisquare

# Observed data
observed = np.array([15, 22, 19, 21, 18, 25])

# Expected data (for a fair die)
expected = np.array([20, 20, 20, 20, 20, 20])

# Chi-square test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)
```
